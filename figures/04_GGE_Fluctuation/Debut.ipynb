```python
import numpy as np
import matplotlib.pyplot as plt

# Fonction pour calculer l'approximation asymptotique de C(N, n)
def C_asymptotic(N, n):
    term1 = N * np.log(N)
    term2 = n * np.log(n)
    term3 = (N - n) * np.log(N - n)
    return np.exp(term1 - term2 - term3)

# Paramètres : N fixé et n allant jusqu'à N-1
N = 1000  # Par exemple, N est fixé à 100
n_values = np.arange(1, N)  # n varie de 1 à N-1

# Calcul des valeurs asymptotiques de C(N, n)
C_values = [C_asymptotic(N, n) for n in n_values]

# Tracer les résultats
plt.figure(figsize=(10, 6))
plt.plot(n_values, C_values, label=f'C(N={N}, n) (approximation)')
plt.title("Comportement asymptotique de C(N, n)")
plt.xlabel("n")
plt.ylabel("C(N, n) (approximation)")
plt.grid(True)
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres : N fixé et n allant jusqu'à N-1
N = 1000  # Par exemple, N est fixé à 100
n_values = np.arange(1, N)  # n varie de 1 à N-1

def factor(n) :
    if n == 1 : return 1 
    return n * factor(n-1)

# Redéfinir la fonction factor pour éviter la récursion excessive
def factor(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Fonction pour calculer C(N, n)
def conb(N, n):
    return factor(N) / (factor(n) * factor(N - n))

# Fonction pour calculer ln(C(N, n)) avec l'approximation
def ln_C_asymptotic(N, n):
    term1 = N * np.log(N)
    term2 = n * np.log(n)
    term3 = (N - n) * np.log(N - n)
    return term1 - term2 - term3

# Fonction pour ln(C(N, n)) exact
def ln_C_exact(N, n):
    return np.log(conb(N, n))

# Recalcul des valeurs exactes et asymptotiques pour ln(C(N, n))
ln_C_asymptotic_values = np.array([ln_C_asymptotic(N, n) for n in n_values])
C_exact_values = np.array([conb(N, n) for n in n_values])
C_asymptotic_values = np.exp(ln_C_asymptotic_values)
ln_C_exact_values = np.array([ln_C_exact(N, n) for n in n_values])


# Tracer les résultats pour comparer

plt.figure(figsize=(12, 6))
plt.plot(N-n_values, C_asymptotic_values - C_exact_values , label="asymptotique - exact", linestyle="--")
plt.title("Comparaison entre C(N, n) exact et asymptotique")
plt.xlabel("N-n")
plt.ylabel("ln(C(N, n))")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
#plt.plot(N-n_values, ln_C_exact_values, label="ln(C(N, n)) exact", linestyle="-")
#plt.plot(N-n_values, ln_C_asymptotic_values, label="ln(C(N, n)) asymptotique", linestyle="--")
plt.plot(N-n_values, ln_C_asymptotic_values - ln_C_exact_values , label="asymptotique - exact", linestyle="--")
plt.title("Comparaison entre ln(C(N, n)) exact et asymptotique")
plt.xlabel("N-n")
plt.ylabel("ln(C(N, n))")
plt.grid(True)
plt.legend()
plt.show()
```
